```python
import tensorflow as tf
imdb = tf.keras.datasets.imdb
```

```python
# Load dataset
num_distinct_words = 5000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=num_distinct_words)
# Retrieve the word index file mapping words to indices
word_index = tf.keras.datasets.imdb.get_word_index()
print(x_train.shape)
print(x_test.shape)
```

```python
# Reverse the word index to obtain a dict mapping indices to words
inverted_word_index = dict((i, word) for (word, i) in word_index.items())
print(len(inverted_word_index.keys()))

print(decoded_sequence)
# Visualize dataset
for i in range(10):
  # Decode the first sequence in the dataset
  decoded_sequence = " ".join(inverted_word_index[i] for i in x_train[i])
  print("Number of words:",len(x_train[0]))
  print(x_train[0])
  print(decoded_sequence)

```

```python
# Preprcess the dataset
max_sequence_length = 300
# Pad all sequences
padded_inputs = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=max_sequence_length, value = 0.0) # 0.0 because it corresponds with <PAD>
padded_inputs_test = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=max_sequence_length, value = 0.0) # 0.0 because it corresponds with <PAD>
```

```python
# Define model
embedding_output_dims = 15
# Define the Keras model
# Create a model
model = tf.keras.models.Sequential([
  tf.keras.layers.Embedding(num_distinct_words, embedding_output_dims, input_length=max_sequence_length),
  tf.keras.layers.LSTM(10),
  tf.keras.layers.Dense(1,activation='sigmoid')
])

# View Model Architecture
model.summary()
```

```python
# Define loss function
# loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
loss_fn = tf.keras.losses.BinaryCrossentropy()
# Compile the model
model.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(padded_inputs, y_train, batch_size=128, epochs=5, validation_split=0.20)
```

```python
# Test the model after training
test_results = model.evaluate(padded_inputs_test, y_test, verbose=False)
print(f'Test results - Loss: {test_results[0]} - Accuracy: {100*test_results[1]}%')
```
